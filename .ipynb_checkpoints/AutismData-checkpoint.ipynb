```python
from scipy.io import arff
import pandas as pd
import seaborn as sns;
import numpy as np
import matplotlib.pyplot as plt

#Loading Data

data = arff.loadarff('Data/Autism-Adult-Data.arff')
autismDf = pd.DataFrame(data[0])
convert_dict = {'A1_Score':int,'A2_Score':int,'A3_Score':int,'A4_Score':int,'A5_Score':int,'A6_Score':int,'A7_Score':int,'A8_Score':int,'A9_Score':int,'A10_Score':int,'age':float,'result':int}
autismDf = autismDf.astype(convert_dict)

for column in autismDf:
    if (autismDf[column].dtype == 'object'):
        #decode object types
        autismDf[column] = autismDf[column].apply(lambda x: x.decode('utf-8'))
        #Standardise data by removing capitals
        autismDf[column] = autismDf[column].apply(lambda x: x.lower())

        
#Preprocessing
#rename columns
autismDf = autismDf.rename(columns={"Class/ASD":"class", "austim":"autism","jundice":"jaundice"})
#remove rows on gender column that aren't f or m
autismDf = autismDf[(autismDf['gender'] == 'f') | (autismDf['gender'] == 'm')]
#Make gender a numerical value
autismDf['gender'] = autismDf['gender'].apply(lambda x: 0 if (x=="f") else 1)    
#remove rows on jundice column that aren't yes or no
autismDf = autismDf[(autismDf['jaundice'] == 'yes') | (autismDf['jaundice'] == 'no')]
#Make jundice a numerical value
autismDf['jaundice'] = autismDf['jaundice'].apply(lambda x: 0 if (x=="no") else 1)    
#remove rows on autism column that aren't yes or no
autismDf = autismDf[(autismDf['autism'] == 'yes') | (autismDf['autism'] == 'no')]
#Make autism a numerical value
autismDf['autism'] = autismDf['autism'].apply(lambda x: 0 if (x=="no") else 1) 
#remove rows on used app before column that aren't yes or no
autismDf = autismDf[(autismDf['used_app_before'] == 'yes') | (autismDf['used_app_before'] == 'no')]
#Make autism a numerical value
autismDf['used_app_before'] = autismDf['used_app_before'].apply(lambda x: 0 if (x=="no") else 1) 
#remove rows on class column that aren't yes or no
autismDf = autismDf[(autismDf['class'] == 'yes') | (autismDf['class'] == 'no')]
#Make class a numerical value
autismDf['class'] = autismDf['class'].apply(lambda x: 0 if (x=="no") else 1) 


# Check for reasonable age
autismDf = autismDf[(autismDf['age'] > 0) & (autismDf['age'] < 110)]
#Convert strings to categoriy codes
autismDf['ethnicity'] = autismDf['ethnicity'].astype('category').cat.codes
autismDf['contry_of_res'] = autismDf['contry_of_res'].astype('category').cat.codes
autismDf['relation'] = autismDf['relation'].astype('category').cat.codes


#drop string columns that are blank
autismDf = autismDf.dropna(subset=['age','gender','ethnicity','jaundice','autism','contry_of_res','result','age_desc','relation','class'])
autismDf = autismDf.drop(['age_desc'], axis=1)

autismDf.head()
```

```python
#Create a correlation matrix to visualise how each feature correlates with the decision value (class)
fig, ax = plt.subplots(figsize=(20,10))
corrMatrix = autismDf.corr(method='pearson')
sns.heatmap(corrMatrix, annot=True, ax=ax)
```

```python
from sklearn import model_selection

#all features except decision
x = autismDf.drop(['class'], axis=1)
#decision value
y = autismDf['class']

xFeatureSelection = autismDf.drop(['jaundice','contry_of_res','used_app_before','gender','class'], axis=1)
```

```python
#Logistic Regression
from sklearn.linear_model import LogisticRegression
logisticReg=LogisticRegression(max_iter=350)
```

```python
#Naive Bayes
from sklearn.naive_bayes import GaussianNB
naiveBayes=GaussianNB()
```

```python
#Decision Tree
from sklearn.tree import DecisionTreeClassifier
decisionTree = DecisionTreeClassifier()
```

```python
#K Nearest Neighbour
from sklearn.neighbors import KNeighborsClassifier

knn = KNeighborsClassifier()
```

```python
#SVM
from sklearn import svm

svm = svm.SVC(kernel='linear')
```

```python
from sklearn.model_selection import cross_validate

def returnCVScores(model):
    #accuracy measures
    scoring = ['accuracy','precision', 'recall', 'f1']
    #get scores with 10-fold cross validation 
    scores = cross_validate(model, x, y, cv=10,
                        scoring=scoring, return_train_score=False)
    #form string with results
    scoresStr = "All Features: \n"
    scoresStr += "Mean Accuracy:" + str(round(scores['test_accuracy'].mean(),3)) + "\n"
    scoresStr += "Mean Precision:" + str(round(scores['test_precision'].mean(),3)) + "\n"
    scoresStr += "Mean Recall:" + str(round(scores['test_recall'].mean(),3)) + "\n"
    scoresStr += "Mean f1:" + str(round(scores['test_f1'].mean(),3)) + "\n"
    
    scoresSelectedFeatures = cross_validate(model, xFeatureSelection, y, cv=10,
                                            scoring=scoring, return_train_score=False)
    scoresStr += "\nSelected Features: \n"
    scoresStr += "Mean Accuracy:" + str(round(scoresSelectedFeatures['test_accuracy'].mean(),3)) + "\n"
    scoresStr += "Mean Precision:" + str(round(scoresSelectedFeatures['test_precision'].mean(),3)) + "\n"
    scoresStr += "Mean Recall:" + str(round(scoresSelectedFeatures['test_recall'].mean(),3)) + "\n"
    scoresStr += "Mean f1:" + str(round(scoresSelectedFeatures['test_f1'].mean(),3)) + "\n"
    
    return scoresStr

#prints all iterations in a line
def prettyPrintIterations(scores):
    iterations = ""
    for score in scores:
        iterations += str(round(score,3)) + ", "
    return iterations

#gets all iterations scores 
def getAllIterationResults(model):
    scoring = ['accuracy','precision', 'recall', 'f1']
    scores = cross_validate(model, x, y, cv=10,
                        scoring=scoring, return_train_score=False)
    scoresStr = "All Features: \n"
    scoresStr += "Mean Accuracy:" + prettyPrintIterations(scores['test_accuracy']) + "\n"
    scoresStr += "Mean Precision:" + prettyPrintIterations(scores['test_precision']) + "\n"
    scoresStr += "Mean Recall:" + prettyPrintIterations(scores['test_recall']) + "\n"
    scoresStr += "Mean f1:" + prettyPrintIterations(scores['test_f1']) + "\n"
    
    scoresSelectedFeatures = cross_validate(model, xFeatureSelection, y, cv=10,
                                            scoring=scoring, return_train_score=False)
    scoresStr += "\nSelected Features: \n"
    scoresStr += "Mean Accuracy:" + prettyPrintIterations(scoresSelectedFeatures['test_accuracy']) + "\n"
    scoresStr += "Mean Precision:" + prettyPrintIterations(scoresSelectedFeatures['test_precision']) + "\n"
    scoresStr += "Mean Recall:" + prettyPrintIterations(scoresSelectedFeatures['test_recall']) + "\n"
    scoresStr += "Mean f1:" + prettyPrintIterations(scoresSelectedFeatures['test_f1']) + "\n"
    
    return scoresStr



    
```

```python
#get mean of all iterations

#Logistic Regression
print('Logistic Regression 10-Fold CV')
print(returnCVScores(logisticReg))

#Naive Bayes
print('\nNaive Bayes 10-Fold CV')
print(returnCVScores(naiveBayes))

#Decision Tree
print('\nDecision Tree 10-Fold CV')
print(returnCVScores(decisionTree))

#K Nearest Neighbour
print('\nK Nearest Neighbour 10-Fold CV')
print(returnCVScores(knn))

#Support Vector Machine
print('\nSupport Vector Machine 10-Fold CV')
print(returnCVScores(svm))
```

```python
#Get all iteration scores

#Logistic Regression
print('Logistic Regression 10-Fold CV Iterations')
print(getAllIterationResults(logisticReg))

#Naive Bayes
print('\nNaive Bayes 10-Fold CV')
print(getAllIterationResults(naiveBayes))

#Decision Tree
print('\nDecision Tree 10-Fold CV')
print(getAllIterationResults(decisionTree))

#K Nearest Neighbour
print('\nK Nearest Neighbour 10-Fold CV')
print(getAllIterationResults(knn))

#Support Vector Machine
print('\nSupport Vector Machine 10-Fold CV')
print(getAllIterationResults(svm))
```
