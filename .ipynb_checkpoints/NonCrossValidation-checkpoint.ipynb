```python
from scipy.io import arff
import pandas as pd
import seaborn as sns;
import numpy as np
import matplotlib.pyplot as plt

#Loading Data

data = arff.loadarff('Data/Autism-Adult-Data.arff')
df = pd.DataFrame(data[0])
convert_dict = {'A1_Score':int,'A2_Score':int,'A3_Score':int,'A4_Score':int,'A5_Score':int,'A6_Score':int,'A7_Score':int,'A8_Score':int,'A9_Score':int,'A10_Score':int,'age':float,'result':int}
df = df.astype(convert_dict)

for column in df:
    if (df[column].dtype == 'object'):
        #decode object types
        df[column] = df[column].apply(lambda x: x.decode('utf-8'))
        #Standardise data by removing capitals
        df[column] = df[column].apply(lambda x: x.lower())

        
#Preprocessing
#rename columns
df = df.rename(columns={"Class/ASD":"class", "austim":"autism","jundice":"jaundice"})
#remove rows on gender column that aren't f or m
df = df[(df['gender'] == 'f') | (df['gender'] == 'm')]
#Make gender a numerical value
df['gender'] = df['gender'].apply(lambda x: 0 if (x=="f") else 1)    
#remove rows on jundice column that aren't yes or no
df = df[(df['jaundice'] == 'yes') | (df['jaundice'] == 'no')]
#Make jundice a numerical value
df['jaundice'] = df['jaundice'].apply(lambda x: 0 if (x=="no") else 1)    
#remove rows on autism column that aren't yes or no
df = df[(df['autism'] == 'yes') | (df['autism'] == 'no')]
#Make autism a numerical value
df['autism'] = df['autism'].apply(lambda x: 0 if (x=="no") else 1) 
#remove rows on used app before column that aren't yes or no
df = df[(df['used_app_before'] == 'yes') | (df['used_app_before'] == 'no')]
#Make autism a numerical value
df['used_app_before'] = df['used_app_before'].apply(lambda x: 0 if (x=="no") else 1) 
#remove rows on class column that aren't yes or no
df = df[(df['class'] == 'yes') | (df['class'] == 'no')]
#Make class a numerical value
df['class'] = df['class'].apply(lambda x: 0 if (x=="no") else 1) 


# Check for reasonable age
df = df[(df['age'] > 0) & (df['age'] < 110)]
#Convert strings to categoriy codes
df['ethnicity'] = df['ethnicity'].astype('category').cat.codes
df['contry_of_res'] = df['contry_of_res'].astype('category').cat.codes
df['relation'] = df['relation'].astype('category').cat.codes


#drop string columns that are blank
df = df.dropna(subset=['age','gender','ethnicity','jaundice','autism','contry_of_res','result','age_desc','relation','class'])
df = df.drop(['age_desc'], axis=1)

df.head()



```
